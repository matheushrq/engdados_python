```python
import pandas as pd

# criando uma série indexada por valores inteiros, iniciando em 0
series = pd.Series([10, 20, 43, 87, 3])
print(series)
```

```python
# criando uma série indexada por letras
series = pd.Series([10, 20, 43, 87, 3], index=['a', 'b', 'c', 'd', 'e'])
print(series)
```

```python
# criando uma série indexada por nomes
series = pd.Series([10, 20, 43, 87, 3], index=['João', 'Maria', 'Pedro', 'José', 'Alex'])
print(series)
```

```python
funcionario = {'João': 1000, 'Maria': 2000, 'Pedro': 43000, 'José': 8700, 'Alex': 3000}
series = pd.Series(funcionario)

# exibindo resumo estatístico da série
series.describe()
```

```python
# ordenando a série pelo índice
series.sort_index(inplace=True) # nesse caso, o índice é o nome, portanto, irá ordenar por ordem alfabética
print(series)
```

```python
# ordenando a série pelo valor
series.sort_values(inplace=True)
print(series)
```

```python
import pandas as pd

notas_Curso = pd.read_csv('Notas_Curso.csv')

# quantidade de linhas por curso
notas_Curso['Curso'].value_counts()

# quantidade de linhas por sexo
notas_Curso['Sexo'].value_counts()

# quantidade de alunos por curso e sexo
notas_Curso.groupby(['Curso', 'Sexo'])['Aluno'].count()

# média de notas por curso e sexo
notas_Curso.groupby(['Curso', 'Sexo']).mean(numeric_only=True)
```

```python
import pandas as pd
notas = pd.read_csv('Notas.csv', sep = ';')
notas

# excluindo linhas que possuem valores nulos somente na coluna D2
notasD2 = notas.dropna(subset=['D2'])
notasD2

# removendo duplicadas
notasD2 = notasD2.drop_duplicates()
notasD2

# gravando um dataframe manipulado em um novo arquivo CSV
notasD2.to_csv('NotasD2.csv', index=False)
```

```python
temp = pd.read_csv('Temperatura.csv', sep=';')
temp.sample(5)

#temp.groupby('cidade')['temperatura'].mean().plot(kind='bar', title='Temperatura média por cidade')

# filtro por cidade e data de coleta
t = temp[(temp.cidade == 'Belo Horizonte') & (temp.data_coleta >= '21/01/2020') & (temp.data_coleta <= '25/01/2020')]
#t

# plotando gráfico de linhas (kind = 'line') com os mesmos dados em BH
t.plot(kind='line', x='data_coleta', y='temperatura', title='Temperatura de BH').set(xlabel='Data', ylabel='Temperatura')
```
