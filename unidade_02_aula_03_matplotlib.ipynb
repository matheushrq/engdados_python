```python
# criando gráfico com 4 conjuntos de linhas
x = [1,2,3,4,5]
y1 = [-3,-2,6,2,1]
y2 = [4,-2,5,8,3]
y3 = [8,5,9,-4,-2]
y4 = [5,8,7,-1,2]

plt.plot(x, y1, 'ro--', x, y2, 'bh-.', x, y3, 'g^:', x, y4, 'y>-')
plt.title('Exemplo');
```

```python
import pandas as pd
import matplotlib.pyplot as plt

weather = pd.read_csv('https://github.com/alanjones2/dataviz/raw/master/londonweather.csv')
weather.head()
```

```python
# obtendo a numeração dos meses avaliados
x_periodo = weather[weather['Year'] == 1957].Month

# obtendo os dados de temperaturas
y_temperatura_1957 = weather[weather['Year'] == 1957].Tmax
y_temperatura_1977 = weather[weather['Year'] == 1977].Tmax
y_temperatura_1997 = weather[weather['Year'] == 1997].Tmax
y_temperatura_2017 = weather[weather['Year'] == 2017].Tmax

y_temperatura_2017
```

```python
plt.plot(x_periodo, y_temperatura_1957);
```

```python
plt.plot(x_periodo, y_temperatura_1957);
plt.title('Temperatura no ano de 1957')
plt.xlabel('Mês');
plt.ylabel('Temperatura');
```

```python
# adicionando elementos ao gráfico

plt.plot(x_periodo, y_temperatura_1957, color='navy', marker='o', linestyle='dashed', linewidth=2, markersize=12);
plt.title('Temperatura no ano de 1957')
plt.xlabel('Mês');
plt.ylabel('Temperatura');
```

```python
# acrescentando GRID

plt.plot(x_periodo, y_temperatura_1957, color='navy', marker='o', linestyle='dashed', linewidth=2, markersize=12);
plt.title('Temperatura no ano de 1957')
plt.xlabel('Mês');
plt.ylabel('Temperatura');
plt.grid(); #axis = 'both', 'x' ou 'y' || axis = posição onde o grid ficará, eixo X ou Y
```

```python
# acrescentando uma anotação

plt.plot(x_periodo, y_temperatura_1957, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12);
plt.annotate('(Tmax = 16.2\n em Maio de 1957)', color='blue', xy = (5, 16.2), xytext=(6, 13), arrowprops=dict(color = 'purple'))
plt.title('Temperatura no ano de 1957')
plt.xlabel('Mês');
plt.ylabel('Temperatura');
plt.grid();
```
