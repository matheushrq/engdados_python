```python
import pandas as pd
import matplotlib.pyplot as plt

weather = pd.read_csv('https://github.com/alanjones2/dataviz/raw/master/londonweather.csv')

weather.sample(5) # 5 linhas de amostra
```

```python
# obtendo a numeração dos meses avaliados
x_periodo = weather[weather['Year'] == 1957].Month

# obtendo os dados de temperaturas
y_temperatura_1957 = weather[weather['Year'] == 1957].Tmax
y_temperatura_1977 = weather[weather['Year'] == 1977].Tmax
y_temperatura_1997 = weather[weather['Year'] == 1997].Tmax
y_temperatura_2017 = weather[weather['Year'] == 2017].Tmax

y_temperatura_1997
```

```python
print(plt.style.available)
```

```python
plt.rcdefaults()
plt.style.use('classic')
plt.plot(x_periodo, y_temperatura_1957, marker='o', label='Tmax em 1957');
plt.annotate('(Tmax = 16.2\n em Maio de 1957)', color='blue', xy = (5, 16.2), xytext=(6, 13), arrowprops=dict(color = 'purple'))
plt.title('Temperatura no ano de 1957')
plt.xlabel('Mês');
plt.ylabel('Temperatura');
#plt.grid();
plt.legend();
```

```python
# gráfico de barras

plt.rcdefaults()
plt.style.use('classic')
plt.bar(x_periodo, y_temperatura_1957, label='Tmax em 1957');

plt.annotate('(Tmax = 23.6\n em Junho de 1957)', color='green', xy = (6, 23), xytext=(1, 18), arrowprops=dict(color = 'red'))

plt.title('Temperatura no ano de 1957')
plt.xlabel('Meses');
plt.ylabel('Temperatura');
plt.grid(axis = 'y');
plt.legend();
```

```python
plt.rcdefaults()

plt.plot(x_periodo, y_temperatura_1957, label='Tmax em 1957', marker='o',);
plt.plot(x_periodo, y_temperatura_1977, label='Tmax em 1977', marker='s');
plt.plot(x_periodo, y_temperatura_1997, label='Tmax em 1997', marker='h');
plt.plot(x_periodo, y_temperatura_2017, label='Tmax em 2017', marker='*');

plt.title('Controle de temperatura máxima dos últimos 60 anos (de 20 em 20 anos)')
plt.ylabel('Temperatura')
plt.xlabel('Meses')
plt.grid()
plt.xticks([]) # oculta rótulo X
plt.legend();
```

```python
# trabalhando com subplots
gridsize = (7,2)
fig = plt.figure(figsize=(9,6))
ax1 = plt.subplot2grid(gridsize, (0,0), colspan=2, rowspan=2) #1957
ax2 = plt.subplot2grid(gridsize, (3,0)) #1977
ax3 = plt.subplot2grid(gridsize, (3,1)) #1997
ax4 = plt.subplot2grid(gridsize, (5,0), colspan=2, rowspan=2) #2017

ax1.set_title('Temperatura por mês em 1957')
ax1.plot(x_periodo, y_temperatura_1957, 'gh--')
ax1.set_xlabel('Meses')
ax1.set_ylabel('1957')
ax1.set_xticks(x_periodo)
ax1.grid()

ax2.set_title('Temperatura por mês em 1977')
ax2.plot(x_periodo, y_temperatura_1977, 'bo--')
ax2.set_xlabel('Meses')
ax2.set_ylabel('1977')
ax2.set_xticks(x_periodo)
ax2.grid()

ax3.set_title('Temperatura por mês em 1997')
ax3.plot(x_periodo, y_temperatura_1997, 'rs--')
ax3.set_xlabel('Meses')
ax3.set_ylabel('1997')
ax3.set_xticks(x_periodo)
ax3.grid()

ax4.set_title('Temperatura por mês em 2017')
ax4.plot(x_periodo, y_temperatura_2017, 'm*-')
ax4.set_xlabel('Meses')
ax4.set_ylabel('2017')
ax4.set_xticks(x_periodo)
ax4.grid()

#plt.tight_layout()
```
